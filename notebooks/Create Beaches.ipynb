```python
import pandas as pd
```

```python
beaches_data = pd.read_csv('data/Beach_Lab_Data.csv')
beaches_data.tail()
```

```python
beaches = beaches_data[['Beach', 'Latitude', 'Longitude']]
beaches = beaches.loc[~beaches['Beach'].isnull()]
beaches
```

```python
beaches = beaches.drop_duplicates()
beaches
```

```python
len(beaches)
```

```python
beaches = beaches.reset_index(drop=True)
beaches
```

```python
beaches.to_csv('data/beaches.csv', index_label='beach_id')
```
