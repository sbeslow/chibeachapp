```python
import pandas as pd
```

```python
lab_results = pd.read_csv('data/Beach_Lab_Data_20170606.csv')
lab_results.head()
```

```python
print("There are %s results, and %s without a location" %
      (len(lab_results), len(lab_results.loc[lab_results['Location'].isnull()])))
```

```python
# Get rid of lab results without locations
lab_results = lab_results.loc[~lab_results['Location'].isnull()]
```

```python
# Check beach names and make sure everything looks Kosher.
print("There are %s beaches, and they are named %s" %
      (len(lab_results['Beach'].unique()),
      ', '.join(lab_results['Beach'].unique())))  
```

```python
# Are there any rows without measurements
lab_results.loc[
                                  lab_results['Culture Sample 1 Reading'].isnull() &
                                  lab_results['Culture Sample 2 Reading'].isnull() &
                                  lab_results['Culture Reading Mean'].isnull() &
                                  lab_results['DNA Sample 1 Reading'].isnull() &
                                  lab_results['DNA Reading Mean'].isnull() &
                                  lab_results['DNA Sample 2 Reading'].isnull()
                                 ]
```

```python
# The city has some datetimes with goofy timestamps.
_ = []
for i, lab_result in enumerate(lab_results.to_dict('records')):
    if not str(lab_result['Culture Sample 1 Timestamp']).lower() == 'nan':
        if (lab_result['Culture Sample 1 Timestamp'].split(' ')[0].split('/')[-1]).startswith('00'):
            print(lab_result)
            lab_result['Culture Sample 1 Timestamp'] = str(lab_result['Culture Sample 1 Timestamp']).replace('0016', '2016')
    _.append(lab_result)
lab_results = pd.DataFrame(_)
lab_results.head()
```

```python
# Same for culture 2
_ = []
for i, lab_result in enumerate(lab_results.to_dict('records')):
    if not str(lab_result['Culture Sample 2 Timestamp']).lower() == 'nan':
        if (lab_result['Culture Sample 2 Timestamp'].split(' ')[0].split('/')[-1]).startswith('00'):
            print(lab_result)
            lab_result['Culture Sample 2 Timestamp'] = str(lab_result['Culture Sample 2 Timestamp']).replace('0016', '2016')
    _.append(lab_result)
lab_results = pd.DataFrame(_)
lab_results.head()
```

```python
# Let's look at the timestamps
lab_results['DNA Sample Timestamp'] = pd.to_datetime(lab_results['DNA Sample Timestamp'])
lab_results['Culture Sample 1 Timestamp'] = pd.to_datetime(lab_results['Culture Sample 1 Timestamp'])
lab_results['Culture Sample 2 Timestamp'] = pd.to_datetime(lab_results['Culture Sample 2 Timestamp'])
lab_results.head()
```

```python
# Are there any samples 
lab_results.loc[
                                  lab_results['Culture Sample 1 Timestamp'].isnull() &
                                  (~lab_results['Culture Sample 2 Timestamp'].isnull())
]
```

```python
# Are there samples with both culture and DNA?
lab_results.loc[
                                  ~lab_results['Culture Reading Mean'].isnull() &
                                  ~lab_results['DNA Reading Mean'].isnull()

                                 ]
```

```python
# Are there samples with only DNA?
lab_results.loc[
                                  lab_results['Culture Reading Mean'].isnull() &
                                  ~lab_results['DNA Reading Mean'].isnull()

                                 ]
```

```python
# Change columns to lowercase and underscores
lab_results.columns = [_.lower().replace(' ', '_') for _ in lab_results.columns]
lab_results.dtypes
```

```python
_ = []
for i, lab_result in enumerate(lab_results.to_dict('records')):
    date, year = None, None
    if str(lab_result['culture_sample_1_timestamp']).lower() != 'nat':
        year = lab_result['culture_sample_1_timestamp'].year
        date = lab_result['culture_sample_1_timestamp'].date()
    if str(lab_result['dna_sample_timestamp']).lower() != 'nat':
        date2 = lab_result['date'] = lab_result['dna_sample_timestamp'].date()
        if not date:
            # print("In it")
            date = date2
            year = date2.year
        elif str(date) != str(date2):
            print('Unmatching dates ' + str(lab_result))
            if date2 > date:
                date = date2
                year = date2.year
    lab_result['year'] = year
    lab_result['date'] = date
    _.append(lab_result)
lab_results = pd.DataFrame(_)
lab_results.head()
```
